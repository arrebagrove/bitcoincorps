```python
# handle_op_checksig using ecdsa.VerifyingKey
import ecdsa
from ecdsa import VerifyingKey, SECP256k1
from ecdsa.ellipticcurve import Point
from ecdsa.ecdsa import Public_key, Signature, generator_secp256k1

def derive_y_from_x(x, is_even):
    """ Derive y point from x point """
    curve = SECP256k1.curve
    # The curve equation over F_p is:
    #   y^2 = x^3 + ax + b
    a, b, p = curve.a(), curve.b(), curve.p()
    alpha = (pow(x, 3, p) + a * x + b) % p
    beta = ecdsa.numbertheory.square_root_mod_prime(alpha, p)
    if (beta % 2) == is_even:
        beta = p - beta
    return beta 

def deserialize_public_key(pk):
    if pk[0] == 4:
        return pk[1:65]
    is_even = pk[0] == 2
    x_hex = pk[1:].hex()
    x = int(x_hex, 16)
    y = derive_y_from_x(x, is_even)
    padded_x = str.format('{:32x}', x)
    padded_y = str.format('{:32x}', y)
    return unhexlify(padded_x)+unhexlify(padded_y)

# This version of handle_op_checksig uses ecdsa VerifyingKey
def handle_op_checksig(stack, tx, input_index):
    # Locking pubkey is atop stack. Grab and interpret.
    pubkey = stack.pop()
    parsed_key = deserialize_public_key(pubkey)

    pk = VerifyingKey.from_string(parsed_key, SECP256k1)

    # Unlocking signature is next. Grab and interpret
    signature = stack.pop()

    # Last byte is the "hash type". Rest are pubkey der.
    sig_der, sig_hash_type = signature[:-1], signature[-1]

    # Get the signature digest (aka hash). Note that Tx.sig_hash 
    # returns the digest/hash in integer format... 
    digest = tx.sig_hash(input_index, sig_hash_type)
    # ... but ecdsa requires the hash to be a string.
    digest = ecdsa.util.number_to_string(digest, SECP256k1.order)

    # PublicKey.verify_digest takes three parameters:
    # (Signature, Digest/Hash (String), Signature decoding function)
    # Since we already have the signature in DER format inside sig_der, 
    # we pass the ecdsa.util.sigdecode_der function as the last parameter.
    sigs_match = pk.verify_digest(sig_der, digest, ecdsa.util.sigdecode_der)

    stack.append(sigs_match)
```
